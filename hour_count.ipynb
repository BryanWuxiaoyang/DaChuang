```python
import pandas as pd
data=pd.read_csv('./stocks_frequency_5.csv',index_col=0)
data.head()
```

```python
data=data.drop_duplicates()
x=data.loc[601318]
print(x)
```

```python
t=x['time']
lst=eval(t)
aa=[i for i in lst if '12-04' in i]
#print(aa)
hours=[i[6:8] for i in aa ]
sr=pd.Series(hours)
vc=sr.value_counts()
print(vc)
```

```python
dvc=dict(vc)
df=pd.DataFrame(list(dvc.items()))
df.columns=['hour','counts']
df
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#sns.reset_defaults()
plt.rcParams['font.sans-serif'] = ['SimHei']  # -
plt.rcParams['axes.unicode_minus'] = False  # '-'
#plt.figure(figsize=(40, 20))
sns.set()
sns.lineplot(x='hour',y='counts',data=df)


plt.show()
```
